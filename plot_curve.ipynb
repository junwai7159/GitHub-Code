```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#epoch,lr,train_loss,train_acc,val_loss,val_acc,precision,recall,fscore,cohen_kappa_score
df = pd.read_csv('./hist.csv')
df = df.iloc[:80]
print(df.columns)

train_loss = np.array(df['train_loss'])
train_acc = np.array(df['train_acc'])
val_loss = np.array(df['val_loss'])
val_acc = np.array(df['val_acc'])
precision = np.array([list(map(float, x[1:-1].split())) for x in list(df['precision'])])
recall = np.array([list(map(float, x[1:-1].split())) for x in list(df['recall'])])
fscore = np.array([list(map(float, x[1:-1].split())) for x in list(df['fscore'])])
cohen_kappa_score = np.array(df['cohen_kappa_score'])
```

```python
# Loss curve
plt.figure()
plt.plot(train_loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.legend()
plt.title('Loss Curve')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.show()

# Accuracy curve
plt.figure()
plt.plot(train_acc)
plt.plot(val_acc)
plt.plot(train_acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.legend()
plt.title('Accuracy Curve')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.show()

# Cohen's Kappa Score
plt.figure()
plt.plot(cohen_kappa_score)
plt.ylabel('Score')
plt.xlabel('Epoch')
plt.title('Cohen\'s Kappa Score')
plt.show()
```

```python
# Overall
plt.figure()
plt.plot(np.sum(precision, axis=1) / 4, label='Precision')
plt.plot(np.sum(recall, axis=1) / 4, label='Recall')
plt.plot(np.sum(fscore, axis=1) / 4, label='F1-Score')
plt.xlabel('Epoch')
plt.ylabel('Percentage')
plt.title('Overall')
plt.legend()
plt.show()

# 0 (Wake)
wake_id = 0
plt.figure()
plt.plot(precision[:, wake_id], label='Precision')
plt.plot(recall[:, wake_id], label='Recall')
plt.plot(fscore[:, wake_id], label='F1-Score')
plt.xlabel('Epoch')
plt.ylabel('Percentage')
plt.title('Wake')
plt.legend()
plt.show()

# 1 (N1/N2)
n1n2_id = 1
plt.figure()
plt.plot(precision[:, n1n2_id], label='Precision')
plt.plot(recall[:, n1n2_id], label='Recall')
plt.plot(fscore[:, n1n2_id], label='F1-Score')
plt.xlabel('Epoch')
plt.ylabel('Percentage')
plt.title('N1/N2')
plt.legend()
plt.show()

# 2 (N3)
n3_id = 2
plt.figure()
plt.plot(precision[:, n3_id], label='Precision')
plt.plot(recall[:, n3_id], label='Recall')
plt.plot(fscore[:, n3_id], label='F1-Score')
plt.xlabel('Epoch')
plt.ylabel('Percentage')
plt.title('N3')
plt.legend()
plt.show()

# 3 (REM)
rem_id = 3
plt.figure()
plt.plot(precision[:, rem_id], label='Precision')
plt.plot(recall[:, rem_id], label='Recall')
plt.plot(fscore[:, rem_id], label='F1-Score')
plt.xlabel('Epoch')
plt.ylabel('Percentage')
plt.title('REM')
plt.legend()
plt.show()
```
